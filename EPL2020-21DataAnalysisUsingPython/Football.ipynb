```python

df['MinsPerMatch']=(df['Mins']/df['Matches']).astype(int)
```

```python
df['GoalsPerMatch']=(df['Goals']/df['Matches']).astype(float)
```

```python
df.head()
```

```python
#Calculate the total goals in 20-21 EPL Season 
total_goals=df['Goals'].sum()

total_goals

print("Total Goals in 20-21 Epl Season := ", total_goals)
```

```python
total_penalty_goals=df['Penalty_Goals'].sum()
print("Total Penalty Goals in 20-21 Epl Season := ", total_penalty_goals)
```

```python
total_penalty_attempts=df['Penalty_Attempted'].sum()


print("Total Penalty attempts in 20-21 Epl Season := ", total_penalty_attempts)
```

```python
plt.figure(figsize=(13,6))
```

```python
penalty_not_scored=df['Penalty_Attempted'].sum()-total_penalty_goals


penalty_not_scored
```

```python
data=[penalty_not_scored,total_penalty_goals]
data

```

```python
#The pie chart is plotted for the specified data. 
#The autopct attribute is used to specify the distance of the text 
#in the slice from the center of the circle.
# The startangle is used to specify the orientation of the chart

labels=['Penalties_Missed','Penalties_Scored']

color=sns.color_palette("Set2")

plt.pie(data,labels=labels,colors=color,autopct='%.9f%%')
plt.show()
```

```python
#PLayer based on the age group 
under_20=df[df['Age']<=20]
#Age of players betn 20 and 25
age_20_25=df[(df['Age']>20) & (df['Age']<=25)]
#Age of players betn 25 and 30 
age_25_30=df[(df['Age']>25) & (df['Age']<=30)]
#players above 30 

age_30=df[df['Age']>30]

age_30.head(10)
```

```python
x=np.array([under_20['Name'].count(),age_20_25['Name'].count(),age_25_30['Name'].count(),age_30['Name'].count()])

labels1=["<=20" ,">20 & <=25" ,">25 & <=30" ,">30"]

plt.title("Total no of players with age group",fontsize=20)
plt.pie(x,labels=labels1,autopct='%.9f%%')
plt.show()
```

```python
#Total no of under 20 players in  each club 
players_20=df[df['Age']<20] 
players_20['Club'].value_counts().plot(kind='bar',color=sns.color_palette('cubehelix'))
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Club',y='Age',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
num_player=df.groupby('Club').size()
data=(df.groupby('Club')['Age'].sum())/num_player


data.sort_values(ascending=False)
```

```python
assists_by_clubs=pd.DataFrame(df.groupby('Club',as_index=False)['Assists'].sum())

sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=assists_by_clubs.sort_values(by="Assists"),palette="Set2")
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Assists',fontsize=20)
```

```python
top_10_assists=df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')

top_10_assists
```

```python
goals_by_clubs=pd.DataFrame(df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=goals_by_clubs.sort_values(by="Goals"),palette='rocket')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"]=(20,8)
plt.title("Plot of Clubs Vs Total Goals ",fontsize=20)
```

```python
#Pie chart for goals wit assists or wiithout assists

plt.figure(figsize=(14,7))
assists=df['Assists'].sum()
data=[total_goals-assists,assists]
labels=['Goals without assists','Goals with assists']
color=sns.color_palette('Set1')

plt.pie(x=data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```

```python
#Top 10 players with most Yellow Cards 

yel=df.sort_values(by='Yellow_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with most Yellow Cards')
c=sns.barplot(x=yel['Name'],y=yel['Yellow_Cards'],label='players',color='yellow')
plt.ylabel('No of yellow cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```

```python
#Top 10 players with most Red Cards 

red=df.sort_values(by='Red_Cards',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with most Red Cards')
c=sns.barplot(x=red['Name'],y=red['Red_Cards'],label='players',color='red')
plt.ylabel('No of red cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```
